```python
with open('./merge/whole_month.data','rb') as f:
    data = pickle.load(f)
    data = np.log10(data + np.ones(data.shape))
M,_,N = data.shape
```

```python
fig,axs = plt.subplots(nrows=math.ceil(M/8), ncols=1, figsize=(20,26))
for i in range(M):
    axs[int(i/8)].plot(data[0,i,:], label=items[i])
for ax in axs:
    ax.legend()
    ax.grid()
plt.show()
```
