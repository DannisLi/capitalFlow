```python
%matplotlib inline
import pickle
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from factorization import Tucker

with open('../merge/whole_month.data','rb') as f:
    cf = pickle.load(f)
```

```python
tucker = Tucker(cf)
tucker.factorize([41,41,36],[1.,1.,1.,2.])
print ('RMSE:', tucker.RMSE())
print ('RRMSE:', tucker.RRMSE())
```

```python
tucker = Tucker(cf,'log')
while True:
    O,D,T,C = tucker.factorize([5,5,4],[1.,1.,1.,2.])
    e = tucker.RRMSE()
    print (e)
    if e<=0.4:
        with open('%.3f.data'%e,'wb') as f:
            pickle.dump((O,D,T,C), f)
```

```python
tucker = Tucker(cf,'sqrt')
tucker.factorize([10,10,9],[1.,1.,1.,2.])
print ('RMSE:', tucker.RMSE())
print ('RRMSE:', tucker.RRMSE())
```
